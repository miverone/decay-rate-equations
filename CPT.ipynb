```python
from decay_rate import dec_rate
import numpy as np
t = np.arange(0,15,0.01)
print(t)
```

```python
rez = 0.1
imz = 0.1
kcosh = 0.5*(1 + rez**2 + imz**2)
kcos = 0.5*(1 - rez**2 - imz**2)
ksinh = 0.5*(2*rez)
ksin = 0.5*(2*imz)
p_plus = dec_rate(t,dm=17.757,dg=(1./1.405 - 1./1.661),gs=1./1.512,D=ksinh,C=kcos,S=ksin,A=kcosh)
p_minus = dec_rate(t,dm=17.757,dg=(1./1.405 - 1./1.661),gs=1./1.512,D=-ksinh,C=kcos,S=-ksin,A=kcosh)
```

```python
from plot_utils import plot_func, plot_osc
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,1, figsize=(30,10))
title = 'Re(z) = {}, Im(z) = {}'.format(rez,imz)
#plot_func(ax,t=t,func=p_plus,xmin=0,xmax=5,ymin=-0.01,ymax=2,col='blue',title=title)
#plot_func(ax,t=t,func=p_minus,xmin=0,xmax=5,ymin=-0.01,ymax=2,col='red',title=title)
plot_osc(ax,t,B_f_t=p_plus,Bbar_f_t=False,B_fbar_t=False,Bbar_fbar_t=p_minus,
             xmin=0,xmax=5,ymin=0,ymax=1,title=title)
plt.show()
```
